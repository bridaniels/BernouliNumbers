### Bernouli Process
- Random Variable Either 1, or 0
    - 1 = Sucess, 0 = Failure
- Probability of Success is Constant 
- Results are Independt of One Another 

#### `X` is *Bernoulli Random Variable*
- `p(0) = P(X=0) = p-1`
- `p(1) = P(X=1) = p`

```python
def perform_bernoulli_trials(trials,probability): 
    n_success = 0 #Success = DEFAULT

    # Trials
    for i in range(trials): 
        # choose random number between zero and one 
        random_number = np.random.random()
        # Less than Probability = Success
        if random_number < probability: 
            n_success += 1
    
    return n_success
```

### Bank with 100 Mortgage Loans
- Possible that within 0:100 will default
- Probability of default is `0.05` 
### Perform 100 Bernoulli Trials 

```python
size = 42
defaults = 1000
loans = 100
prob = 0.05
```

```python
# Seed Random Number Generator - initializes random number generator 
np.random.seed(42)
n_defaults = np.empty(defaults)

for i in range(defaults): 
    n_defaults[i] = perform_bernoulli_trials(loans, prob)


# Plottings
plt.figure(figsize=(8,6), frameon=False)

plt.hist(n_defaults, density=True, rwidth=0.9)

plt.xlabel('Number of Defaults/100 Loans', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.title('Defaults Probaility', fontsize=20)

plt.show()
```

## How Likely Is Bank to Default? 


```python
def emperical_cumulative_distribution_function(data):
    n = len(data)

    #x-data
    x = np.sort(data)
    #y-data
    y = np.arange(1,1+n) / n 

    return x, y 
```

```python
x, y = emperical_cumulative_distribution_function(n_defaults)



plt.figure(figsize=(10,6))
ax = plt.axes()
ax.set_facecolor("LightGrey")
plt.grid(color='white')

plt.plot(x,y, marker='.', linestyle='none', color='b')

plt.xlabel('Number of Defaults/100', fontsize=12)
plt.ylabel('Cumulative Distribution Function (CDF)', fontsize=12)
plt.title('Probability of Bank to Lose Money', fontsize=20)

plt.show()
```

```python
# Probability of X + Defaults 
default = 10 
n_lose_money = np.sum(n_defaults >= default)
print('Probability of the Bank Losing Money: {}'.format(n_lose_money/len(n_defaults)))
```
